```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(11,21)

# scatter plot
plt.scatter(x,y,c='b')
```

```python
plt.plot(x,y)
```

```python
y = x*x
plt.scatter(x,y,c='g')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("graph")
plt.savefig('Test.png')
```

```python
y = x*x
plt.plot(x,y,'r--')
```

```python
# creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

```python
x = np.arange(5,20)
y = 2 * x + 5
plt.plot(x,y,'r')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("equation of line")
```

```python
x = np.arange(0, 2 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 
plt.plot(x, y) 
```

```python
x = np.arange(0, 8*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin, 'b--')
plt.title("sin curve")
plt.subplot(2,1,2)
plt.plot(x, y_cos, 'r--')
plt.title("cos curve")
```

```python
# Bar plot

x = [2,8,10] 
y = [11,16,9]  
plt.bar(x, y, color = 'b')  
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  
```

```python
x = [2,8,10] 
y = [11,16,9]
x2 = [3,9,11] 
y2 = [6,15,7]
plt.bar(x, y, color = 'b') 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  
```

```python
# histogram
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

```python
# box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
```

```python
# pie chart
# data
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0.1, 0)  # explodes 1st slice, 3rd slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
