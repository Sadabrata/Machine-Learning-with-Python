```python
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
# tip is the dependent feature
df.head()
```

```python
df.dtypes
```

```python
df.corr()
# correlation is not calculated for object data type
```

```python
sns.heatmap(df.corr())
```

## JoinPlot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation.

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

## Pair plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

## Distplot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

# Catagorical Plots

```python
# Count plot
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
# Bar plot
sns.barplot(x='sex',y='total_bill',data=df)
# shown wrt catagorical feature
```

```python
# box plot
sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
# Violin Plot
# Violin plot helps us to see both the distribution of data in terms of Kernel density estimation
# and the box plot
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
